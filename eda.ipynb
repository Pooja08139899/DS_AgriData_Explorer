```python
#Read data from dataset
file_path="ICRISAT.csv"
df=pd.read_csv(file_path)
df.head()
```

```python
df.isnull()
```

```python
df.dropna(axis=1,how='all',inplace=True)
```

```python
state_rice = df.groupby("State Name")["RICE PRODUCTION (1000 tons)"].sum().reset_index()

# Get top 7 rice-producing states
top7_rice_states = state_rice.sort_values(by="RICE PRODUCTION (1000 tons)", ascending=False).head(7)

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(top7_rice_states["State Name"], top7_rice_states["RICE PRODUCTION (1000 tons)"], color=["skyblue","red","orange","pink","yellow","green","violet"])
plt.xlabel("State")
plt.ylabel("Rice Production (1000 tons)")
plt.title("Top 7 Rice Producing States in India")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
state_wheat = df.groupby("State Name")["WHEAT PRODUCTION (1000 tons)"].sum().reset_index()

# Get top 5 wheat-producing states
top5_wheat_states = state_wheat.sort_values(by="WHEAT PRODUCTION (1000 tons)", ascending=False).head(5)

# --- Bar chart ---
plt.figure(figsize=(10,6))
plt.bar(top5_wheat_states["State Name"], top5_wheat_states["WHEAT PRODUCTION (1000 tons)"], color=["violet","pink","green","yellow","orange"])
plt.xlabel("State")
plt.ylabel("Wheat Production (1000 tons)")
plt.title("Top 5 Wheat Producing States in India")
plt.xticks(rotation=30, ha="right")
plt.show()

# --- Pie chart ---
plt.figure(figsize=(8,8))
plt.pie(top5_wheat_states["WHEAT PRODUCTION (1000 tons)"], 
        labels=top5_wheat_states["State Name"], 
        autopct="%.2f%%", 
        startangle=140, 
        colors=["gold", "lightcoral", "lightskyblue", "yellowgreen", "violet"])
plt.title("Percentage Share of Top 5 Wheat Producing States")
plt.show()
```

```python
state_oilseeds = df.groupby("State Name")["OILSEEDS PRODUCTION (1000 tons)"].sum().reset_index()

# Get top 5 oilseed-producing states
top5_oilseeds_states = state_oilseeds.sort_values(by="OILSEEDS PRODUCTION (1000 tons)", ascending=False).head(5)

# --- Scatter plot ---
plt.figure(figsize=(10,6))
plt.scatter(top5_oilseeds_states["State Name"], 
            top5_oilseeds_states["OILSEEDS PRODUCTION (1000 tons)"], 
            color="green", s=150, marker="o")

# Add labels for each point
for i, row in top5_oilseeds_states.iterrows():
    plt.text(row["State Name"], row["OILSEEDS PRODUCTION (1000 tons)"]+50, 
             f'{int(row["OILSEEDS PRODUCTION (1000 tons)"])}', 
             ha='center', fontsize=10)

plt.xlabel("State")
plt.ylabel("Oilseeds Production (1000 tons)")
plt.title("Top 5 Oilseeds Producing States in India")
plt.xticks(rotation=30, ha="right")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# Group by state and sum sunflower production across all years & districts
state_sunflower = df.groupby("State Name")["SUNFLOWER PRODUCTION (1000 tons)"].sum().reset_index()

# Get top 7 sunflower-producing states
top7_sunflower_states = state_sunflower.sort_values(by="SUNFLOWER PRODUCTION (1000 tons)", ascending=False).head(7)

# --- Donut chart ---
plt.figure(figsize=(8,8))
sizes = top7_sunflower_states["SUNFLOWER PRODUCTION (1000 tons)"]
labels = top7_sunflower_states["State Name"]

wedges, texts, autotexts = plt.pie(
    sizes,
    labels=labels,
    autopct="%.1f%%",
    startangle=140,
    pctdistance=0.85,
    colors=plt.cm.tab20c.colors
)

# Draw circle for donut hole
centre_circle = plt.Circle((0,0),0.70,fc="white")
plt.gca().add_artist(centre_circle)

plt.title("Top 7 Sunflower Producing States in India (Donut Chart)")
plt.tight_layout()
plt.show()
```

```python
# Group by year and sum sugarcane production across all states
sugarcane_yearly = df.groupby("Year")["SUGARCANE PRODUCTION (1000 tons)"].sum().reset_index()

# Get last 50 years of data
last50 = sugarcane_yearly.sort_values(by="Year", ascending=True).tail(50)

# --- Line plot ---
plt.figure(figsize=(12,6))
plt.plot(last50["Year"], last50["SUGARCANE PRODUCTION (1000 tons)"], 
         marker="o", color="green", linewidth=2)

plt.xlabel("Year")
plt.ylabel("Sugarcane Production (1000 tons)")
plt.title("India's Sugarcane Production in the Last 50 Years")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Group by year and sum rice & wheat production across all states
yearly_production = df.groupby("Year")[["RICE PRODUCTION (1000 tons)", "WHEAT PRODUCTION (1000 tons)"]].sum().reset_index()

# Get last 50 years of data
last50 = yearly_production.sort_values(by="Year", ascending=True).tail(50)

# --- Stacked Area Chart ---
plt.figure(figsize=(12,6))
plt.stackplot(last50["Year"], 
              last50["RICE PRODUCTION (1000 tons)"], 
              last50["WHEAT PRODUCTION (1000 tons)"], 
              labels=["Rice", "Wheat"], 
              colors=["skyblue", "orange"], alpha=0.8)

plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.title("Rice vs Wheat Production in India (Last 50 Years)")
plt.legend(loc="upper left")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Filter data for West Bengal
wb_data = df[df["State Name"] == "West Bengal"]

# Group by district and sum rice production
wb_rice = wb_data.groupby("Dist Name")["RICE PRODUCTION (1000 tons)"].sum().reset_index()

# Sort by production
wb_rice = wb_rice.sort_values(by="RICE PRODUCTION (1000 tons)", ascending=False)

# --- Bar chart ---
plt.figure(figsize=(12,6))
plt.bar(wb_rice["Dist Name"], wb_rice["RICE PRODUCTION (1000 tons)"], color="teal")
plt.xticks(rotation=90)
plt.xlabel("Districts of West Bengal")
plt.ylabel("Rice Production (1000 tons)")
plt.title("Rice Production by Districts of West Bengal")
plt.tight_layout()
plt.show()
```

```python

# Filter data for Uttar Pradesh
up_data = df[df["State Name"] == "Uttar Pradesh"]

# Group by year and sum wheat production across all UP districts
up_wheat = up_data.groupby("Year")["WHEAT PRODUCTION (1000 tons)"].sum().reset_index()

# Get Top 10 years with highest production
top10_years = up_wheat.sort_values(by="WHEAT PRODUCTION (1000 tons)", ascending=False).head(10)
top10_years = top10_years.sort_values(by="Year")  # sort by year for logical x-axis order

# --- Bar chart ---
plt.figure(figsize=(10,6))
plt.bar(top10_years["Year"].astype(str), top10_years["WHEAT PRODUCTION (1000 tons)"], color="green")
plt.xlabel("Year")
plt.ylabel("Wheat Production (1000 tons)")
plt.title("Top 10 Wheat Production Years in Uttar Pradesh")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Compute total millet production (Sorghum + Pearl Millet + Finger Millet)
df["MILLET PRODUCTION (1000 tons)"] = (
    df["SORGHUM PRODUCTION (1000 tons)"] +
    df["PEARL MILLET PRODUCTION (1000 tons)"] +
    df["FINGER MILLET PRODUCTION (1000 tons)"]
)

# Group by year and sum across states
millet_yearly = df.groupby("Year")["MILLET PRODUCTION (1000 tons)"].sum().reset_index()

# Get last 50 years
last50 = millet_yearly.sort_values(by="Year", ascending=True).tail(50)

# --- Line Chart ---
plt.figure(figsize=(12,6))
plt.plot(last50["Year"], last50["MILLET PRODUCTION (1000 tons)"], 
         marker="o", color="purple", linewidth=2)

plt.xlabel("Year")
plt.ylabel("Millet Production (1000 tons)")
plt.title("India's Millet Production (Last 50 Years)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python
# Group by state and sum Kharif & Rabi sorghum production
sorghum_state = df.groupby("State Name")[["KHARIF SORGHUM PRODUCTION (1000 tons)", 
                                            "RABI SORGHUM PRODUCTION (1000 tons)"]].sum().reset_index()

# Plot grouped bar chart
x = np.arange(len(sorghum_state["State Name"]))  # positions
width = 0.35  # bar width

plt.figure(figsize=(14,8))
plt.bar(x - width/2, sorghum_state["KHARIF SORGHUM PRODUCTION (1000 tons)"], 
        width, label="Kharif", color="goldenrod")
plt.bar(x + width/2, sorghum_state["RABI SORGHUM PRODUCTION (1000 tons)"], 
        width, label="Rabi", color="teal")

plt.xticks(x, sorghum_state["State Name"], rotation=90)
plt.xlabel("State (Region)")
plt.ylabel("Sorghum Production (1000 tons)")
plt.title("Sorghum Production (Kharif vs Rabi) by Region")
plt.legend()
plt.tight_layout()
plt.show()





```

```python
# Group by state and sum groundnut production
state_groundnut = df.groupby("State Name")["GROUNDNUT PRODUCTION (1000 tons)"].sum().reset_index()

# Get Top 7 states
top7_groundnut = state_groundnut.sort_values(by="GROUNDNUT PRODUCTION (1000 tons)", ascending=False).head(7)

# --- Horizontal Bar Chart ---
plt.figure(figsize=(10,6))
plt.barh(top7_groundnut["State Name"], top7_groundnut["GROUNDNUT PRODUCTION (1000 tons)"], color="sandybrown")
plt.xlabel("Groundnut Production (1000 tons)")
plt.ylabel("State")
plt.title("Top 7 Groundnut Producing States in India")
plt.gca().invert_yaxis()  # highest at top
plt.tight_layout()
plt.show()
```

```python
# Group by state: sum production and average yield
state_soybean = df.groupby("State Name").agg({
    "SOYABEAN PRODUCTION (1000 tons)": "sum",
    "SOYABEAN YIELD (Kg per ha)": "mean"
}).reset_index()

# Get top 5 states by production
top5_soybean = state_soybean.sort_values(by="SOYABEAN PRODUCTION (1000 tons)", ascending=False).head(5)

# --- Plot (Bar + Line) ---
fig, ax1 = plt.subplots(figsize=(10,6))

# Bar chart for production
ax1.bar(top5_soybean["State Name"], top5_soybean["SOYABEAN PRODUCTION (1000 tons)"], 
        color="goldenrod", alpha=0.7, label="Production (1000 tons)")
ax1.set_xlabel("State")
ax1.set_ylabel("Soybean Production (1000 tons)", color="goldenrod")
ax1.tick_params(axis="y", labelcolor="goldenrod")

# Line chart for yield efficiency
ax2 = ax1.twinx()
ax2.plot(top5_soybean["State Name"], top5_soybean["SOYABEAN YIELD (Kg per ha)"], 
         color="darkgreen", marker="o", linewidth=2, label="Yield (Kg/ha)")
ax2.set_ylabel("Yield (Kg per ha)", color="darkgreen")
ax2.tick_params(axis="y", labelcolor="darkgreen")

# Title
plt.title("Top 5 Soybean Producing States and Yield Efficiency")
fig.tight_layout()
plt.show()
```

```python
# Group by state and sum oilseed production
state_oilseeds = df.groupby("State Name")["OILSEEDS PRODUCTION (1000 tons)"].sum().reset_index()

# Sort by production (descending)
state_oilseeds = state_oilseeds.sort_values(by="OILSEEDS PRODUCTION (1000 tons)", ascending=False)

# --- Horizontal Bar Chart ---
plt.figure(figsize=(12,7))
plt.barh(state_oilseeds["State Name"], state_oilseeds["OILSEEDS PRODUCTION (1000 tons)"], color="olive")
plt.xlabel("Oilseed Production (1000 tons)")
plt.ylabel("State")
plt.title("Oilseed Production in Major States of India")
plt.gca().invert_yaxis()  # highest at top
plt.tight_layout()
plt.show()
```

```python
df1 = df[[
    "RICE AREA (1000 ha)", "RICE PRODUCTION (1000 tons)",
    "WHEAT AREA (1000 ha)", "WHEAT PRODUCTION (1000 tons)",
    "MAIZE AREA (1000 ha)", "MAIZE PRODUCTION (1000 tons)"
]]

# --- Scatter Plot ---
plt.figure(figsize=(10,6))

plt.scatter(df1["RICE AREA (1000 ha)"], df["RICE PRODUCTION (1000 tons)"], 
            color="skyblue", alpha=0.5, label="Rice")
plt.scatter(df1["WHEAT AREA (1000 ha)"], df["WHEAT PRODUCTION (1000 tons)"], 
            color="orange", alpha=0.5, label="Wheat")
plt.scatter(df1["MAIZE AREA (1000 ha)"], df["MAIZE PRODUCTION (1000 tons)"], 
            color="green", alpha=0.5, label="Maize")

plt.xlabel("Area Cultivated (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.title("Impact of Area Cultivated on Production (Rice, Wheat, Maize)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Group by state and take average yield
yield_state = df.groupby("State Name")[["RICE YIELD (Kg per ha)", "WHEAT YIELD (Kg per ha)"]].mean().reset_index()

# Plot side-by-side bar chart
x = np.arange(len(yield_state["State Name"]))  # state positions
width = 0.35  # bar width

plt.figure(figsize=(14,6))
plt.bar(x - width/2, yield_state["RICE YIELD (Kg per ha)"], width, label="Rice Yield", color="teal")
plt.bar(x + width/2, yield_state["WHEAT YIELD (Kg per ha)"], width, label="Wheat Yield", color="goldenrod")

plt.xticks(x, yield_state["State Name"], rotation=90)
plt.xlabel("States")
plt.ylabel("Yield (Kg per ha)")
plt.title("Rice vs Wheat Yield Across States")
plt.legend()
plt.tight_layout()
plt.show()
```
